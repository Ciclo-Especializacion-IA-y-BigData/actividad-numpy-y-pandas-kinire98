#  **Pandas Series**

##  ¬øQu√© es una *Series*?

Una **Series** de Pandas es como una **columna de una tabla** o de una hoja de c√°lculo.
Es una **estructura unidimensional** (una sola dimensi√≥n) que puede almacenar **datos de cualquier tipo**: n√∫meros, cadenas, valores booleanos, etc.

Puedes imaginarla como una lista con nombre o una columna de Excel con √≠ndices.

---

##  Crear una *Series* desde una lista

üìò **Ejemplo:**

````python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
````

Cada valor tiene una **etiqueta o √≠ndice** (label).
Por defecto, Pandas asigna √≠ndices num√©ricos empezando desde 0.

---

##  **√çndices o etiquetas (Labels)**

Si no especificas nada, los elementos se etiquetan autom√°ticamente como:

```
0, 1, 2, 3, ...
```

Estos √≠ndices te permiten **acceder a un valor concreto**.

üìò **Ejemplo: acceder al primer valor**

````python
print(myvar[0])
````

---

##  **Crear tus propias etiquetas**

Puedes asignar tus **propias etiquetas** (√≠ndices personalizados) con el argumento `index`.

üìò **Ejemplo:**

````python
import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
````


Y ahora puedes acceder por etiqueta en lugar de por posici√≥n:

````python
print(myvar["y"])
````

Otra forma muy √∫til de crear una Series es a partir de un **diccionario de Python**.
Las **claves** del diccionario se convierten autom√°ticamente en los **√≠ndices**.

üìò **Ejemplo:**

````python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
````

##  **Seleccionar elementos concretos**

Puedes elegir **solo algunas claves del diccionario** indicando los nombres que quieres conservar en el argumento `index`.

üìò **Ejemplo:**

````python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
````

 **De Series a DataFrame**

En Pandas, los conjuntos de datos suelen tener **varias columnas** (no solo una).
Cuando combinamos varias *Series* con distintos nombres de columna, obtenemos un **DataFrame**, que es una **tabla completa de datos**.

üìò **Ejemplo:**

````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
````

En este ejemplo, **cada columna es una Series**, y el **DataFrame** las une en una sola estructura.

#  Introducci√≥n a los DataFrames en Pandas

## ¬øQu√© es un DataFrame?

Un **DataFrame** de **Pandas** es una estructura de datos **bidimensional** (2D), similar a una **tabla** con filas y columnas.
Piensa en √©l como una hoja de c√°lculo de Excel, pero **dentro de Python**, con superpoderes para an√°lisis de datos.

üîπ Cada columna puede tener un tipo distinto (n√∫meros, texto, booleanos‚Ä¶).

üîπ Cada fila tiene un **√≠ndice** (por defecto empieza en 0).

üîπ Es la base del **an√°lisis de datos** y de la **preparaci√≥n de datasets** en IA.

---

##  Crear un DataFrame


````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# Cargar los datos en un objeto DataFrame
df = pd.DataFrame(data)

print(df)
````

Observa:

* Cada clave del diccionario (`"calories"`, `"duration"`) se convierte en una **columna**.
* Cada elemento de la lista es un **valor** de fila.

---

## Localizar filas con `loc`

Puedes acceder a filas espec√≠ficas usando el **√≠ndice** o el **nombre de la fila** (si lo has definido).

### Ejemplo 1: Obtener la primera fila

````python
print(df.loc[0])
````

**Devuelve una Series**, no un DataFrame, porque solo hay una fila.

---

üìò **Ejemplo 2: devolver varias filas**

````python
print(df.loc[[0, 1]])
````

En este caso, al usar una **lista de √≠ndices**, Pandas devuelve otro **DataFrame**.

---

## üè∑Ô∏è **√çndices con nombre (Named Indexes)**

Puedes asignar **nombres personalizados** a las filas mediante el argumento `index`.

üìò **Ejemplo:**

````python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index=["day1", "day2", "day3"])

print(df)
````

Ahora las filas tienen **etiquetas significativas** en lugar de n√∫meros.

---

## üìç **Acceder a una fila por nombre**

Una vez asignados nombres, puedes usar `.loc[]` con esos nombres.

üìò **Ejemplo:**

````python
print(df.loc["day2"])
````

En proyectos reales de *Machine Learning*, los datos no se escriben a mano, sino que se **cargan desde archivos**: CSV, Excel, JSON, SQL, etc.

Pandas facilita esto con funciones como `read_csv()` o `read_excel()`.

üìò **Ejemplo:**

Vamos a usar el dataset real de empleados disponible en GitHub:

````python
import pandas as pd

# Cargar el dataset desde la URL
url = "https://raw.githubusercontent.com/sivabalanb/Data-Analysis-with-Pandas-and-Python/refs/heads/master/employees.csv"
df = pd.read_csv(url)

# Mostrar las primeras filas
print(df.head())
````

#  Lecci√≥n: Leer archivos CSV con Pandas

##  ¬øQu√© es un archivo CSV?

Los archivos **CSV** (*Comma Separated Values*) son una forma **sencilla y universal** de guardar grandes conjuntos de datos.
Cada l√≠nea representa una **fila**, y las columnas est√°n separadas por **comas (,)**.

üîπ Son archivos de texto plano.
üîπ Se pueden abrir con Excel, Google Sheets o cualquier editor de texto.
üîπ Se usan constantemente en proyectos de **Big Data** y **Machine Learning**.

---

##  Cargar un archivo CSV con Pandas

Pandas facilita enormemente leer un archivo CSV.
Vamos a cargar un dataset llamado `data.csv`.

````python
import pandas as pd

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df.to_string())
````


 `to_string()` muestra **todo el DataFrame completo** (aunque sea grande).

---

###  Sin `to_string()`

Si no usamos `to_string()`, Pandas mostrar√° **solo las primeras y √∫ltimas 5 filas**, para no saturar la pantalla:

````python
import pandas as pd

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df)
````


---

## ‚öôÔ∏è Controlar cu√°ntas filas se muestran

Pandas tiene una configuraci√≥n interna llamada `pd.options.display.max_rows`, que define el n√∫mero m√°ximo de filas que se muestran por defecto.

---

### üîç Ver cu√°ntas filas m√°ximas se muestran

````python
import pandas as pd

print(pd.options.display.max_rows)
````


‚û°Ô∏è Esto significa que, si tu DataFrame tiene m√°s de 60 filas, solo se mostrar√°n **las primeras y las √∫ltimas 5**.

### üîß Cambiar el l√≠mite de filas visibles

Si quieres mostrar **todas** las filas de un DataFrame, puedes aumentar este l√≠mite:

````python
import pandas as pd

pd.options.display.max_rows = 9999   # Aumentamos el m√°ximo

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/Breast%20Cancer%20Dataset.csv"

df = pd.read_csv(url)

print(df)
````


‚ö†Ô∏è **Cuidado:** Si tu archivo tiene miles de filas, puede tardar en mostrarse todo.

## üí° Ejemplo de uso real

Existen muchas APIS gratuitas de las que podemos obtener datos, algunas requieren registro previo para tener un api key:

https://free-apis.github.io/

Sup√≥n que tienes una API que devuelve datos meteorol√≥gicos en formato JSON.
Con Pandas puedes cargar esos datos directamente y analizarlos.

````python
import requests
import pandas as pd

# 1. Definir los par√°metros para la API
# Coordenadas de Le√≥n, Espa√±a
latitud = 42.60
longitud = -5.57

# Rango de fechas para los datos hist√≥ricos (ej. todo el a√±o 2023)
fecha_inicio = "2023-01-01"
fecha_fin = "2023-12-31"

# Variables diarias que queremos obtener:
# - temperature_2m_max: Temperatura m√°xima diaria a 2 metros del suelo.
# - temperature_2m_min: Temperatura m√≠nima diaria.
# - precipitation_sum: Suma total de precipitaci√≥n (lluvia, nieve...) en mm.
# - wind_speed_10m_max: Velocidad m√°xima del viento a 10 metros en km/h.
parametros_diarios = "temperature_2m_max,temperature_2m_min,precipitation_sum,wind_speed_10m_max"

# 2. Construir la URL para la API de datos hist√≥ricos
# ¬°OJO! La URL es diferente a la de pron√≥stico: usa "archive-api.open-meteo.com"
url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitud}&longitude={longitud}"
    f"&start_date={fecha_inicio}&end_date={fecha_fin}"
    f"&daily={parametros_diarios}"
    f"&timezone=Europe/Madrid"  # Especificar la zona horaria es una buena pr√°ctica
)

try:
    # 3. Realizar la petici√≥n GET a la API
    respuesta = requests.get(url)
    respuesta.raise_for_status()  # Lanza un error si la petici√≥n no fue exitosa (ej. error 4xx o 5xx)

    # 4. Convertir la respuesta a formato JSON
    datos_json = respuesta.json()

    # 5. Extraer los datos diarios y guardarlos en un DataFrame de pandas
    datos_diarios = datos_json.get('daily')

    if datos_diarios:
        df_historico = pd.DataFrame(datos_diarios)

        # Convertir la columna 'time' a un formato de fecha m√°s manejable
        df_historico['time'] = pd.to_datetime(df_historico['time'])

        # Renombrar columnas para mayor claridad
        df_historico.rename(columns={
            'time': 'Fecha',
            'temperature_2m_max': 'Temp Max (¬∞C)',
            'temperature_2m_min': 'Temp Min (¬∞C)',
            'precipitation_sum': 'Precipitaci√≥n (mm)',
            'wind_speed_10m_max': 'Viento Max (km/h)'
        }, inplace=True)

        print(f"--- Datos meteorol√≥gicos hist√≥ricos de Le√≥n para 2023 ---")
        print(f"Se han obtenido {len(df_historico)} registros.")

        print("\nPrimeros 5 d√≠as del a√±o:")
        print(df_historico.head())

        print("\n√öltimos 5 d√≠as del a√±o:")
        print(df_historico.tail())

    else:
        print("No se encontraron datos diarios en la respuesta de la API.")

except requests.exceptions.RequestException as e:
    print(f"Error al conectar con la API: {e}")
````


En IA, este flujo es t√≠pico para alimentar **modelos predictivos en tiempo real**.

---

##  **1. ¬øQu√© son los ‚Äúdatos err√≥neos‚Äù?**

Los **datos err√≥neos** no siempre est√°n vac√≠os o mal formateados.
Pueden ser valores **que no tienen sentido l√≥gico o estad√≠stico** dentro del contexto del dataset.

üîπ Ejemplo cl√°sico:
En un registro de entrenamientos, la **duraci√≥n media** suele ser entre 30 y 60 minutos.
Si aparece un valor de **450 minutos**, es probable que sea un **error humano** (un ‚Äú0‚Äù extra).

---

##  **2. Dataset de ejemplo**

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

print(df.to_string())
````


üí° Todo parece correcto en la caracter√≠stica *Duration* salvo la duraci√≥n: 450 minutos (7,5 horas) no tiene sentido.

---

## üß© **3. Sustituir un valor err√≥neo (m√©todo directo)**

Podemos **corregir manualmente** un valor aislado con `.loc[]`:

````python
df.loc[7, "Duration"] = 45

print(df.head(10))
````


‚úÖ Esto reemplaza el valor err√≥neo por **45** en la fila 7.

üìå √ötil cuando el dataset es peque√±o o los errores son muy concretos.

---

##  **4. Crear reglas autom√°ticas para detectar y corregir**

En datasets grandes no podemos corregir valor a valor.
Podemos definir **l√≠mites v√°lidos (umbrales)** y aplicar reglas autom√°ticas.

Por ejemplo, si todas las duraciones deben estar entre 30 y 120 minutos:

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

for x in df.index:
    if df.loc[x, "Duration"] > 120:
        df.loc[x, "Duration"] = 120

print(df.head(10))
````


üí° `.loc[]` act√∫a como una lupa sobre el DataFrame y permite modificar datos directamente.

---

## üîÅ **5. Alternativa m√°s ‚Äúpandasiana‚Äù (sin bucles)**

Pandas permite hacerlo sin necesidad de recorrer fila a fila:

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df.loc[df["Duration"] > 120, "Duration"] = 120


print(df.head(10))
````


‚úÖ Mismo resultado, pero m√°s r√°pido y legible.

---

##  **6. Eliminar filas con valores err√≥neos**

En algunos casos no conviene reemplazar, sino **eliminar completamente** los registros que rompen la coherencia.

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df = df.drop(df[df["Duration"] > 120].index)

print(df.head(10))

````


üí° Esta opci√≥n se usa cuando:

* No sabemos cu√°l ser√≠a el valor correcto.
* Los datos err√≥neos son pocos y no afectan el an√°lisis.

---

##  **7. Comprobar la correcci√≥n**



````python
print(df["Duration"].describe())
````


 Ya no hay valores an√≥malos por encima de 120.

3. ¬øC√≥mo podr√≠as aplicar reglas similares para detectar errores en otras columnas (por ejemplo, `Pulse` o `Calories`)?

## ‚úÖ **Resumen**

| Situaci√≥n                | Acci√≥n                        | Ejemplo                                                 |
| ------------------------ | ----------------------------- | ------------------------------------------------------- |
| Error aislado y evidente | Corregir directamente         | `df.loc[7, "Duration"] = 45`                            |
| Valores fuera de rango   | Reemplazar con l√≠mite o media | `df.loc[df["Duration"] > 120, "Duration"] = 120`        |
| Datos sin sentido        | Eliminar fila                 | `df.drop(df[df["Duration"] > 120].index, inplace=True)` |
| Validaci√≥n general       | Revisar estad√≠sticas          | `df.describe()`                                         |

#  **Eliminaci√≥n de duplicados en Pandas**

---

##  **1. Qu√© son los duplicados**

En los datasets reales, es habitual encontrar **filas duplicadas**, es decir, **registros repetidos exactamente iguales**.

Esto puede suceder cuando:

* Se combinan fuentes de datos distintas.
* Se guardan varias veces las mismas mediciones.
* Hay errores de registro o importaci√≥n.

üîπ **Ejemplo del dataset de entrenamientos**
Observa las filas 11 y 12: son id√©nticas.

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

print(df.to_string())
````

---

##  **2. Detectar duplicados con `duplicated()`**

Pandas ofrece el m√©todo `duplicated()` para detectar filas repetidas.

````python
print(df.duplicated())
````


üí° **Interpretaci√≥n:**

* `True` ‚Üí la fila es **un duplicado**.
* `False` ‚Üí la fila es **√∫nica**.

---

##  **3. Ver solo las filas duplicadas**

Para ver las filas duplicadas directamente:

````python
print(df[df.duplicated()])
````


---

##  **4. Eliminar duplicados con `drop_duplicates()`**

Para **eliminar filas repetidas**, usamos `drop_duplicates()`.

````python
df.drop_duplicates(inplace=True)
````


üìå Con `inplace=True` modificamos el DataFrame original directamente.
Sin ese argumento, el m√©todo devuelve una **copia** limpia.

---

## ‚öôÔ∏è **5. Comprobaci√≥n final**


````python
print(df.duplicated().sum())
````


‚úÖ Ya no hay filas duplicadas.

---

##  **6. Eliminar duplicados seg√∫n columnas espec√≠ficas**

A veces queremos considerar duplicado solo si coinciden ciertas columnas (por ejemplo, mismas fechas y pulsaciones):

````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/data.csv'
df = pd.read_csv(url)

df.drop_duplicates(subset=["Date", "Pulse"], inplace=True)

print(df.duplicated().sum())
````


As√≠ mantenemos la fila m√°s reciente o completa, y eliminamos las redundantes.

---

## üß† **7. Evitar duplicados al combinar datos**

Cuando unimos datasets con `concat()` o `merge()`, es buena pr√°ctica revisar despu√©s si se han duplicado registros:

````python
df1 = pd.DataFrame({
    "Nombre": ["Ana", "Luis", "Mar√≠a"],
    "Edad": [23, 30, 25]
})

df2 = pd.DataFrame({
    "Nombre": ["Luis", "Mar√≠a", "Pedro"],
    "Edad": [30, 25, 28]
})

df = pd.concat([df1, df2])

print(df)

df.drop_duplicates(inplace=True)

print("\n Eliminamos duplicados\n")

print(df)
````


üí¨ **Consejo:** Incorpora siempre un paso de verificaci√≥n de duplicados en tu flujo de limpieza (*pipeline*).

---

## ‚úÖ **Resumen**

| Tarea               | M√©todo                             | Descripci√≥n                    |
| ------------------- | ---------------------------------- | ------------------------------ |
| Detectar duplicados | `df.duplicated()`                  | Devuelve `True/False` por fila |
| Mostrar duplicados  | `df[df.duplicated()]`              | Filtra las filas repetidas     |
| Eliminar duplicados | `df.drop_duplicates(inplace=True)` | Borra las filas duplicadas     |
| Basado en columnas  | `subset=["col1", "col2"]`          | Compara solo esas columnas     |



#  **Lecci√≥n: Correlaciones de Datos en Pandas**

---

##  **1. ¬øQu√© es una correlaci√≥n?**

En an√°lisis de datos e inteligencia artificial, una **correlaci√≥n** mide **c√≥mo se relacionan dos variables**:
si aumentan o disminuyen juntas, o si no guardan relaci√≥n alguna.

Pandas nos permite calcular correlaciones de forma muy sencilla con el m√©todo `.corr()`.

---

##  **2. Dataset de ejemplo**

Usaremos el dataset de entrenamientos `training_data.csv` con m√°s datos:


````python
url = 'https://raw.githubusercontent.com/carlostessier/DataSets/refs/heads/main/csv/training_data.csv'
df = pd.read_csv(url)

print(df.head())
````


---

##  **3. Calcular la correlaci√≥n**

Pandas tiene un m√©todo muy √∫til: **`df.corr()`**, que calcula autom√°ticamente las correlaciones entre todas las columnas num√©ricas del DataFrame.

````python
print(df.corr(numeric_only=True))
````


---

##  **4. Interpretaci√≥n de los resultados**

Cada n√∫mero indica **la fuerza y direcci√≥n de la relaci√≥n** entre dos variables.
El valor var√≠a de **-1 a 1**:

| Valor           | Interpretaci√≥n                | Significado                      |
| --------------- | ----------------------------- | -------------------------------- |
| **1.0**         | Correlaci√≥n perfecta positiva | Ambas variables aumentan juntas  |
| **0.8 a 0.6**   | Correlaci√≥n fuerte            | Si una sube, la otra suele subir |
| **0.4 a 0.2**   | Correlaci√≥n d√©bil             | Relaci√≥n poco fiable             |
| **0.0**         | Sin correlaci√≥n               | No hay relaci√≥n lineal           |
| **-0.6 a -1.0** | Correlaci√≥n negativa          | Cuando una sube, la otra baja    |

---

##  **5. Ejemplo visual: buena vs mala correlaci√≥n**

| Ejemplo                           | Descripci√≥n                                                  | Interpretaci√≥n        |
| --------------------------------- | ------------------------------------------------------------ | --------------------- |
| üîπ **Duration ‚Äì Calories:** 0.92  | Cuanto m√°s se entrena, m√°s calor√≠as se queman.               | Buena correlaci√≥n ‚úÖ   |
| üîπ **Duration ‚Äì Maxpulse:** 0.009 | No hay relaci√≥n directa entre la duraci√≥n y el pulso m√°ximo. | Mala correlaci√≥n ‚ùå    |
| üîπ **Pulse ‚Äì Maxpulse:** 0.78     | A mayor pulso medio, mayor pulso m√°ximo.                     | Correlaci√≥n fuerte üí™ |

---

##  **6. Visualizaci√≥n de correlaciones**

Podemos mostrar la matriz de correlaciones como un **mapa de calor (heatmap)** para interpretarla f√°cilmente.

````python
import matplotlib.pyplot as plt
import seaborn as sns

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlaci√≥n de variables del entrenamiento")
plt.show()
````


 Esto genera un gr√°fico donde los colores indican la fuerza de la relaci√≥n:

* üî¥ **Rojo fuerte ‚Üí correlaci√≥n positiva**
* üîµ **Azul fuerte ‚Üí correlaci√≥n negativa**

---

## üí¨ **7. Aplicaci√≥n en Inteligencia Artificial**

En IA y *Machine Learning*, la correlaci√≥n es fundamental porque:

* Permite **detectar variables redundantes o poco relevantes**.
* Ayuda a **reducir la dimensionalidad** del modelo (quedarse solo con las variables √∫tiles).
* Sirve como paso previo a t√©cnicas como **regresi√≥n lineal o PCA (an√°lisis de componentes principales)**.

üí° Ejemplo:

> Si ‚ÄúDuration‚Äù y ‚ÄúCalories‚Äù tienen una correlaci√≥n de 0.92, podr√≠as usar una para predecir la otra con un modelo de regresi√≥n.
